```python
#Consulta para analizar el valor de las ventas totales por región
#Para esto se realizaron uniones entre tablas mediante la instrucción INNER JOIN
cur.execute("""WITH SCA AS (
    SELECT 
        c.CATEGORIA, 
        s.SEGMENTO,
        s.ATTR1, s.ATTR2, s.ATTR3, s.FORMATO
    FROM CATEGORY c
    INNER JOIN SEGMENT s 
        ON s.ID_DE_CATEGORIA = c.ID_DE_CATEGORIA
)
SELECT 
    s.FORMATO,
    s.SEGMENTO,
    sa.REGION,
    SUM(sa.TOTAL_DE_UNIDADES_VENDIDAS) AS unidades_vendida,
    SUM(sa.VALOR_TOTAL_DE_VENTAS) AS ventas_totales 
FROM SCA s
INNER JOIN PRODUCT p 
    ON (s.ATTR1 = p.ATTR1) 
   AND (s.ATTR2 = p.ATTR2) 
   AND (s.ATTR3 = p.ATTR3) 
   AND (s.FORMATO = p.FORMATO)
INNER JOIN SALES sa 
    ON sa.CODIGO = p.CODIGO
    GROUP BY 
    sa.REGION, s.SEGMENTO,   s.FORMATO
ORDER BY 
    SUM(sa.VALOR_TOTAL_DE_VENTAS) DESC;
HAVING SUM(sa.VALOR_TOTAL_DE_VENTAS)<1000     
     """)
result1 = cur.fetchall()
result1
```

```python
#Consulta para analizar el valor de las ventas totales por región
#Para esto se realizaron uniones entre tablas mediante la instrucción INNER JOIN
cur.execute("""WITH SCA AS (
    SELECT 
        c.CATEGORIA, s.SEGMENTO, s.ATTR1, s.ATTR2, s.ATTR3, s.FORMATO
    FROM CATEGORY c
    INNER JOIN SEGMENT s 
        ON s.ID_DE_CATEGORIA = c.ID_DE_CATEGORIA)
    SELECT 
       s.FORMATO, s.SEGMENTO, sa.REGION, SUM(sa.TOTAL_DE_UNIDADES_VENDIDAS) AS unidades_vendida,
       SUM(sa.VALOR_TOTAL_DE_VENTAS) AS ventas_totales 
    FROM SCA s
    INNER JOIN PRODUCT p 
        ON (s.ATTR1 = p.ATTR1) 
       AND (s.ATTR2 = p.ATTR2) 
       AND (s.ATTR3 = p.ATTR3) 
       AND (s.FORMATO = p.FORMATO)
    INNER JOIN SALES sa 
        ON sa.CODIGO = p.CODIGO
    GROUP BY 
        sa.REGION
    HAVING SUM(sa.VALOR_TOTAL_DE_VENTAS)>2000000
    ORDER BY 
      ventas_totales DESC;   """)
result1 = cur.fetchall()
result1
```

```python
#Consulta para analizar el valor de las ventas totales por segmento
#Para esto se realizaron uniones entre tablas mediante la instrucción INNER JOIN
cur.execute("""WITH SCA AS (
    SELECT 
        c.CATEGORIA, s.SEGMENTO, s.ATTR1, s.ATTR2, s.ATTR3, s.FORMATO
    FROM CATEGORY c
    INNER JOIN SEGMENT s 
        ON s.ID_DE_CATEGORIA = c.ID_DE_CATEGORIA)
SELECT 
    s.FORMATO, s.SEGMENTO, SUM(sa.TOTAL_DE_UNIDADES_VENDIDAS) AS unidades_vendida,
    SUM(sa.VALOR_TOTAL_DE_VENTAS) AS ventas_totales 
FROM SCA s
INNER JOIN PRODUCT p 
    ON (s.ATTR1 = p.ATTR1) 
   AND (s.ATTR2 = p.ATTR2) 
   AND (s.ATTR3 = p.ATTR3) 
   AND (s.FORMATO = p.FORMATO)
INNER JOIN SALES sa 
    ON sa.CODIGO = p.CODIGO
    GROUP BY 
    s.SEGMENTO
HAVING
    SUM(sa.VALOR_TOTAL_DE_VENTAS)<500000
ORDER BY 
    ventas_totales DESC; """)
result1 = cur.fetchall()
result1
```

```python
#Consulta para analizar el valor de las ventas totales por fecha (por meses en específico)
#Para esto se realizaron uniones entre tablas mediante la instrucción INNER JOIN
cur.execute("""WITH SCA AS (
    SELECT 
        c.CATEGORIA, 
        s.SEGMENTO,
        s.ATTR1, s.ATTR2, s.ATTR3, s.FORMATO
    FROM CATEGORY c
    INNER JOIN SEGMENT s 
        ON s.ID_DE_CATEGORIA = c.ID_DE_CATEGORIA
)
SELECT 
    sa.REGION,
    ca.MONTH,
    ca.YEAR,
    SUM(sa.VALOR_TOTAL_DE_VENTAS) AS ventas_totales 
FROM SCA s
INNER JOIN PRODUCT p 
    ON (s.ATTR1 = p.ATTR1) 
   AND (s.ATTR2 = p.ATTR2) 
   AND (s.ATTR3 = p.ATTR3) 
   AND (s.FORMATO = p.FORMATO)

INNER JOIN SALES sa 
    ON sa.CODIGO = p.CODIGO
INNER JOIN CALENDAR ca
    ON REPLACE(sa.SEMANAS, ' ', '') = REPLACE(ca.WEEK, ' ', '')
WHERE sa.REGION = 'TOTAL AUTOS SCANNING MEXICO'
GROUP BY 
   ca.YEAR, ca.MONTH
ORDER BY 
    ca.MONTH,ca.YEAR DESC;
     """)
result1 = cur.fetchall()
result1
```
